<b>According to the Pearson's correlation matrix "Incidence" and "Slope" apear to have the highest correlation. <br><brThe null hypothesis states: The slope of normal and abnormal classifications is the same.  With a confidence interval of .05 the ttest is used to determine validity.

```python
print("Mean of total slope population: ", np.mean(df['slope']))
print("St Dev of total slope population: ", np.std(df['slope']))
```

```python
norm_slope = df[df['class'] == 'Normal']['slope']
abn_slope = df[df['class'] == 'Abnormal']['slope']
n_norm = len(norm_slope)
n_abn = len(abn_slope)

print("Degree of freedom = ",n_norm + n_abn-2)
```

```python
n = len(df['slope'])
tn = n-1
critical_val = t(tn).ppf(0.95)
print("Critical Value: ",critical_val)
```

```python
lower_boundry = df.slope.mean() - critical_val / np.sqrt(n) * df.slope.std()
print("Lower Boundry: ", lower_boundry)
```

```python

var_norm = norm_degree.var(ddof=1)
var_abn = abn_degree.var(ddof=1)

#std deviation
s = np.sqrt((var_norm + var_abn)/2)
s
```

```python
     #Store values needed for manual formula calculations
mn_abn = np.mean(norm_slope)
mn_norm = np.mean(abn_slope)
std_abn = np.std(norm_slope)
std_norm = np.std(abn_slope)

print("Uninsured - Mean: ", mn_abn, ", Standard Deviation: ", std_abn, ", Size: ", n_abn)
print("Insured - Mean: ", mn_norm, ", Standard Deviation: ", std_norm, ", Size: ", n_norm)
```

```python
sp = np.sqrt(((n_norm-1)*std_norm**2+(n_abn-1)*std_abn**2)/(n_norm+n_norm-2)) # pooled standard deviation
ttest = (mn_norm-mn_abn)/(sp*np.sqrt(1/n_norm+1/n_abn)) # t statistic
print("Test Statistic: ",ttest)
```

```python
degrees = n_norm + n_abn-2
p_value = 2*t.cdf(ttest, degrees)
print("Probability Value: ",p_value)
```

#### Performing a KFold split of 5 results in same accuracy

```python
def cv_score(clf, x, y, score_func=accuracy_score):
    result = 0
    nfold = 6
    for train, test in KFold(nfold).split(x): # split data into train/test groups, 5 times
        clf.fit(x[train], y[train]) # fit
        result += score_func(clf.predict(x[test]), y[test]) # evaluate score function on held-out data
    return result / nfold # average
```

```python
clf = LogisticRegression(solver='lbfgs')
score = cv_score(clf, xtrain, ytrain)
print(score)
print("Initial accuracy score for Logistic Regression: ", accuracy_score(clf.predict(xtest), ytest))
```

#### Performing a GridSearch to optimize the C parameter...

```python
from sklearn.model_selection import GridSearchCV
def cv_optimize(clf, parameters, xtrain, ytrain, n_folds=5):
    gs = GridSearchCV(clf, param_grid=parameters, cv=n_folds)
    gs.fit(xtrain, ytrain)
    print("BEST PARAMS", gs.best_params_)
    best = gs.best_estimator_
    return best
```

```python
parameters = {"C": [0.0001, 0.001, 0.1, 0.5, 1, 10]}
print(cv_optimize(clf, parameters, xtrain, ytrain, n_folds=5))
```
